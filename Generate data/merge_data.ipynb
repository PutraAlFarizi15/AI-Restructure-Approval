```python
import pandas as pd
```

```python
df_data = pd.read_csv('../data/Dummy Data AI Restructure Approval rev7.csv')
df_data.info()
```

```python
df_data = df_data.rename(columns={'occupation': 'debtor_occupation', 
                                  'age' : 'debtor_age'})
df_data.info()
```

```python
df_data = df_data.drop(['debtor_name', 'debtor_nik', 'debtor_address', 'debtor_number',
                        'debtor_company'], axis=1)
df_data.info()
```

```python
df_profil = pd.read_csv('df_prof.csv')
df_profil.info()
```

```python
df_profil = df_profil.drop(['debtor_occupation', 'debtor_age'], axis=1)
df_profil.info()
```

```python
# Menggabungkan kolom dari kedua DataFrame
df_gabung = pd.concat([df_profil, df_data], axis=1)
df_gabung.info()
```

```python
# Save DataFrame to file CSV
df_gabung.to_csv('../data/Dummy Data AI Restructure Approval rev7.csv', index=False)
```

```python
idx_id = 3777
df_gabung = df_gabung[:idx_id]
df_gabung.info()
```
