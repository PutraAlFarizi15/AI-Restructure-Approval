```python
import pandas as pd
```

```python
df = pd.read_csv('../data/Dataset AI Restructure Approval rev7 05102023.csv')
df.info()
```

```python
df_profil = df[df.columns[:14]]
df_profil
```

```python
output = ['old_monthly_payments', 'loan_amount','tenor', 'monthly_payments']
df_output = df[output]
df_output
```

```python
df_init_tenor = df.drop(output, axis=1)
df.info()
```

```python
df_recommendation_tenor= pd.concat([df_init_tenor, df_output], axis=1)
df_recommendation_tenor.info()
```

```python
df_recommendation_tenor.to_csv('../data/Dataset Recommendation Tenor and Monthly Payments 05102023.csv', index=False)
```

```python
loan = ['monthly_income', 'debt', 'dti', 'monthly_expenses',
        'net_income','previous_credit_monthly_income',
        'financial_changes', 'business_conditions', 'asset_value',
        'interest_rate', 'collateral','old_collateral', 'ltv',
        'tenor', 'loan_amount']
df_loan = df[loan]
df_loan.info()
```

```python
df_request_loan= pd.concat([df_profil, df_loan], axis=1)
df_request_loan.info()
```

```python
df_request_loan.to_csv('../data/Dataset Request Loan 05102023.csv', index=False)
```
